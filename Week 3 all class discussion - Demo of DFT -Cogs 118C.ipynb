```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile
```

```python
# DFT of a song
Fs, aud = wavfile.read('HW1_audio.wav')
x = aud[:,0]
N = len(x)
t = np.arange(0,N/Fs,1/Fs)

x = np.sin(2*np.pi*t*1000)

print(N)
```

```python
plt.plot(t, x);
plt.xlim([0,.01])
plt.xlabel('Time (s)')
```

```python
X = np.fft.fft(x) # An algorithm for calculating the DFT (discrete Fourier transform)
f = np.fft.fftfreq(N, d=1/Fs)
```

```python
X[1], np.conj(X[-1])
np.abs(X[1]), np.abs(X[-1])
```

```python
plt.plot( np.abs(X)**2)
plt.yscale('log')
```

```python
plt.plot(f, np.abs(X)**2,'o-')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power (dB)')
```
