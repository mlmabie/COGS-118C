```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def boxcar(t, N):
    # Boxcar function made by summing N sinusoids
    y = 0;
    for n in range(N):
        y = y+(1/(2*n+1))*np.sin(t*(2*n+1))
    return y
```

```python
boxcar(0.1, 100)
```

```python
t = np.arange(-2,2,0.01)
plt.plot(t, boxcar(2*np.pi*t, N=20))
```

```python
# Sin is an odd function
plt.plot(t, np.sin(2*np.pi*t))
plt.plot(t, -np.sin(2*np.pi*(-t)))
plt.grid()
```

```python
# cos is an even function
plt.plot(t, np.cos(2*np.pi*t))
plt.plot(t, np.cos(2*np.pi*(-t)), '.')
plt.grid()
```

```python
plt.plot(t, np.sin(t**2))
```

```python
# Check the orthonormality of sin and cos
t = np.arange(0,2*np.pi,0.001)
f = np.sin(t)   # n=1
g = np.sin(t)   # m=1

plt.plot(t, f)
plt.plot(t, g)
plt.plot(t, f*g)
plt.grid()
```

```python
np.mean(f*g)
```
