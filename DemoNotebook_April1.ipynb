```python
f = np.sin(2*np.pi*t)
```

```python
plt.plot(t, f)
plt.grid()
```

```python
f = np.sin(2*np.pi*t/5) + np.cos(2*np.pi*t/np.sqrt(5))

plt.figure(figsize=(15,5))
plt.plot(t, f)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Oscillation f(t)')
```

$\sin(2 \pi t/5) + cos(2 \pi t/\sqrt{5}) = \sin(2 \pi t+2\pi T/5) + cos(2 \pi t/\sqrt{5} +2\pi T/\sqrt{5})$

$2\pi T/5 = 2\pi k$ where $k$ is an integer, $k\in Z$
